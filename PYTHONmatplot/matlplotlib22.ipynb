```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Colored Scatterplots
iris=pd.read_csv("iris.csv")
```

```python
iris.sample(5)
```

```python
#plot size
plt.figure(figsize=(9,7)) #width and length
plt.scatter(iris["SepalLengthCm"],iris["PetalLengthCm"],c=iris["Species"],cmap="jet",alpha=0.8)
plt.xlabel("Speal Length")
plt.ylabel("Petal Length")
plt.colorbar()
```

```python
fig, ax =plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(10,6)) #2*2

ax[0].scatter(batters["avg"],batters["strike_rate"],color="red")
ax[1].scatter(batters["avg"],batters["runs"])

ax[0].set_title("avg vs strike rate")
ax[0].set_xlabel("avg")
ax[0].set_ylabel("strike rate")

ax[1].set_title("Avg vs Runs")
ax[1].set_xlabel("Avg")
ax[1].set_ylabel("Runs")
```

#3D scatter PLOTS

```python
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
x
```

```python
xx,yy=np.meshgrid(x,y)
```

```python
z=xx**2 +yy**2
```

```python
fig=plt.figure(figsize=(15,10))
ax=plt.subplot(projection="3d")
ax.plot_surface(xx,yy,z,cmap="viridis")
```

```python
z=np.sin(xx) + np.cos(yy)
fig=plt.figure(figsize=(15,10))
ax=plt.subplot(projection="3d")
p=ax.plot_surface(xx,yy,z,cmap="viridis")
fig.colorbar(p)
```

```python
#contour plot
z=xx**2 +yy**2
fig=plt.figure(figsize=(15,10))
ax=plt.subplot()
p=ax.contour(xx,yy,z)
fig.colorbar(p)
```

```python
#contour plot
z=xx**2 +yy**2
fig=plt.figure(figsize=(15,10))
ax=plt.subplot()
p=ax.contourf(xx,yy,z)
fig.colorbar(p)
```

```python
z=np.sin(xx) + np.cos(yy)
fig=plt.figure(figsize=(15,10))
ax=plt.subplot()
p=ax.contourf(xx,yy,z,cmap="viridis")
fig.colorbar(p)
```
