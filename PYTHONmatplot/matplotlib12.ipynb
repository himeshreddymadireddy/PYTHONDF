```python
batsman=pd.read_csv("sharma-kohli.csv")
```

```python
plt.plot(batsman["index"],batsman["V Kohli"])
```

```python
plt.plot(batsman["index"],batsman["V Kohli"])
plt.plot(batsman["index"],batsman["RG Sharma"])
```

```python
#labels title
plt.plot(batsman["index"],batsman["V Kohli"])
plt.plot(batsman["index"],batsman["RG Sharma"])

plt.title("Rohit Sharma VS VK Kohli")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
```

```python
#color , line witdh and b ,marker(size)
plt.plot(batsman["index"],batsman["V Kohli"],color="#3efe00")
plt.plot(batsman["index"],batsman["RG Sharma"],color="black")

plt.title("Rohit Sharma VS VK Kohli")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
```

```python
#line witdh and b ,marker(size)
plt.plot(batsman["index"],batsman["V Kohli"],color="#3efe00",linestyle="solid",linewidth=3)
plt.plot(batsman["index"],batsman["RG Sharma"],color="black",linestyle=":",linewidth=2)

plt.title("Rohit Sharma VS VK Kohli")
plt.xlabel("Season")
plt.ylabel("Runs Scored")

#'-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
```

```python
# markers are called points
#marker(size)
plt.plot(batsman["index"],batsman["V Kohli"],color="#3efe00",linestyle="solid",linewidth=1,marker="o",markersize=3)
plt.plot(batsman["index"],batsman["RG Sharma"],color="black",linestyle="solid",linewidth=1,marker="P")

plt.title("Rohit Sharma VS VK Kohli")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
```

```python
# markers are called points
#marker(size)
plt.plot(batsman["index"],batsman["V Kohli"],color="#3efe00",linestyle="solid",linewidth=1,marker="o",markersize=3,label="virat")
plt.plot(batsman["index"],batsman["RG Sharma"],color="black",linestyle="solid",linewidth=1,marker="P",label="rg sharma")

plt.title("Rohit Sharma VS VK Kohli")
plt.xlabel("Season")
plt.ylabel("Runs Scored")

plt.legend(loc="upper right")
```

```python
#limiting axes
price=[48000,54000,57000,48000,46000,4500000]
year=[2015,2016,2017,2018,2019,2021]
plt.plot(year,price)
plt.xlim(2017,2019)
plt.ylim(0,75000)
```

```python
#grid
plt.plot(batsman["index"],batsman["V Kohli"],color="#3efe00",linestyle="solid",linewidth=1,marker="o",markersize=3,label="virat")
plt.plot(batsman["index"],batsman["RG Sharma"],color="black",linestyle="solid",linewidth=1,marker="P",label="rg sharma")

plt.title("Rohit Sharma VS VK Kohli")
plt.xlabel("Season")
plt.ylabel("Runs Scored")

plt.legend(loc="upper right")
plt.grid()
```

```python
#explicit show of graphs compulsory in py charm and it is 
plt.plot(batsman["index"],batsman["V Kohli"],color="#3efe00",linestyle="solid",linewidth=1,marker="o",markersize=3,label="virat")
plt.plot(batsman["index"],batsman["RG Sharma"],color="black",linestyle="solid",linewidth=1,marker="P",label="rg sharma")

plt.title("Rohit Sharma VS VK Kohli")
plt.xlabel("Season")
plt.ylabel("Runs Scored")

plt.legend(loc="upper right")
plt.grid()
plt.show()
```

```python
df=pd.read_csv("batsman_season_record.csv")
df
```

```python
plt.bar(df["batsman"],df["2015"])
plt.bar(df["batsman"],df["2016"])
#one upon another
```

```python
plt.bar(df["batsman"],df["2015"],width=0.5)
```

```python
plt.bar(np.arange(df.shape[0])-0.2,df["2015"],width=0.2)
plt.bar(np.arange(df.shape[0]),df["2016"],width=0.2)
plt.bar(np.arange(df.shape[0])+0.2,df["2017"],width=0.2)
plt.show()

#mutliple bar charts
```

```python
plt.bar(np.arange(df.shape[0])-0.2,df["2015"],width=0.2,color=("#03e6ff"))
plt.bar(np.arange(df.shape[0]),df["2016"],width=0.2)
plt.bar(np.arange(df.shape[0])+0.2,df["2017"],width=0.2,color=("#ffae00"))



plt.xticks(np.arange(df.shape[0]),df["batsman"])
plt.xlabel("Batsman")
plt.ylabel("Scores")
plt.title("highest scores of 2016,2017,2018")
plt.show()
```

```python
# stagerring of x labels changing
children=[10,20,40,10,20]
colors=["red red red red","blue blue blue blue","yellow yellow yellow yellow","pink pink pink pink","yellow yellow yellow yellow"]
plt.bar(colors,children,color="red")
plt.xticks(rotation="vertical")
```

```python
#stacked bar chart
plt.bar(df["batsman"],df["2017"],label="2015")
plt.bar(df["batsman"],df["2016"],bottom=df["2017"],label="2016")
plt.bar(df["batsman"],df["2016"],bottom=df["2016"]+df["2017"],label="2017")
plt.title("2017 vs 2016 vs 2015")
plt.legend()
plt.show()

#ADD UP ALL THE ABOVE
```
