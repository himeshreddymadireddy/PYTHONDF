```python
point=pd.DataFrame({
    "1st point":[(3,4),(12,3),(2,1),(3,200),(2,3),(12,1)],
    "2nd point":[(1,100),(2,3),(12,1),(3,4),(21,3),(2,1)]
})
```

```python
point
```

```python
def euceldian(row):
    
    pt_A=row["1st point"]
    pt_B=row["2nd point"]
    
    return ((pt_A[0]-pt_B[1])**2 +(pt_A[1]-pt_B[1])**2)**0.5
```

```python
point.apply(euceldian,axis=1)
```
