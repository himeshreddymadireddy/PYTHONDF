```python
#apply(series+datframes)
points_df = pd. DataFrame (
    {
        '1st point': [ (3,4), (-6,5), (0,0), (-10,1) , (4,5) ],
        '2nd point': [(-3,4), (0,0), (2,2), (10, 10), (1,1) ]
    }
)
points_df
```

```python
def euclidean_distance(row):
    pt_A=row["1st point"]
    pt_B=row["2nd point"]
    
    return ((pt_A[0]-pt_B[1])**2)+((pt_B[1]-pt_A[1])**2)**0.5
    
```

```python
points_df.apply(euclidean_distance,axis=1)
```
