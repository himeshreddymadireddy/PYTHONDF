```python
#list vs numpy array
a1=[i for i in range(10000000)]
b1=[i for i in range(10000000,20000000)]
```

```python
c=[]

import time
#dynamic array # adress is stored
start=time.time()
for i in range(len(a1)):
    c.append(a1[i]+b1[i])
print(time.time()-start)
```

```python
import numpy as np
a=np.arange(10000000,dtype=np.int32) #uses c type array static ,size fixed
# directly stored
b=np.arange(10000000,20000000,dtype=np.int32)
start=time.time()
c=a+b
print(time.time()-start)
```
