# SERIES 1D data

```python
import numpy as np
import pandas as pd
```

# series from list

```python
country=["India","Nepal","Bangladesh","Nepal","Srilanka"]
pd.Series(country)
```

```python
Runs=[100,29,123,34]
RUNS=pd.Series(Runs)
```

```python
#custom index
subjects=["maths","english","hindi","science"]
pd.Series(Runs,index=subjects)
```

```python
#setting a name 
marks=pd.Series(Runs,index=subjects,name="marks of virat")
```

```python
marks
```

# series from dict

```python
marks={
    "maths":98,
    "science":88,
    "hindi":90,
    "english":81,
}
```

```python
marks_series=pd.Series(marks,name="virat marks")
```

```python
marks_series
```
